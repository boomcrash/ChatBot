<center><h1>PROCESAMIENTO DE LENGUAJE NATURAL<h1/> <center/>
<center><img src="https://assets.website-files.com/5ad5cd6cfa9d76ac6aacb916/5fb131a4de4f3f590099dbc9_Untitled%20design.png"><center/>

```
words=[]
all_words=[]
tags= []
aux= []
auxA= []
auxB= []
training=[]
exit=[]
 
try:
  with open("/content/chatbot/Entrenamiento/brain.pickle","rb") as pickleBrain:
    all_words,tags,training,exit=pickle.load(pickleBrain)
except:
  for intent in database['intents']:
    #recorremos los tags
    for pattern in intent['patterns']:
      #separamos una frase en palabras
      auxWords=nltk.word_tokenize(pattern)
      #guardamos las palabras
      auxA.append(auxWords)
      auxB.extend(auxWords)
      #guardamos los tags
      aux.append(intent['tag'])
  #simbolos a ignorar
  ignore_words=['?','!','.',',','¿',"'","$","·","-","_","&","%","/","(",")", "=","#"]
  for w in auxB:
    if w not in ignore_words:
      words.append(w)
  #truco para evitar repetidos
  words=sorted(set(words))
  print(words)    
  tags=sorted(set(aux))
  print(tags)
  #convertimos a minuscula
  all_words=[stemmer.stem(w.lower()) for w in words]
  print("SIN ORDENAR")
  print(len(all_words))
  #ordenamos la lista
  ##print("ORDENADO")
  all_words=sorted(list(set(all_words)))
  ##print(all_words)
  #ordenamos los tags
  tags=sorted(tags)
  ##print("TAGS ORDENADO")
  ##print(tags)
  training=[]
  exit=[]
  #creamos una salida falsa
  null_exit=[0 for _ in range(len(tags))]
  #recorremos el auxiliar de palabras
  for i, document in enumerate(auxA):
    bucket=[]
    #hacemos minuscula y quitamos signos
    auxWords=[stemmer.stem(w.lower())for w in document if w!="?"]
    #recorremos las palabras
    for w in all_words:
      if w in auxWords:
        bucket.append(1)
      else:
        bucket.append(0)
    exit_row=null_exit[:]
    exit_row[tags.index(aux[i])]=1
    training.append(bucket)
    exit.append(exit_row)
    ##print(training)
    ##print(exit)
     #pasamos la lista del entrenamiento a un arreglo numpy
  training=numpy.array(training)
  ##print(training)
  exit=numpy.array(exit)
  ##print(exit)
  #crear un archivo pickle para almacenar los datos entrenados
  with open("/content/chatbot/Entrenamiento/brain.pickle","wb") as pickleBrain:
    pickle.dump((all_words,tags,training,exit),pickleBrain)
```

<center><h1> CREAMOS LA RED NEURONAL<h1/><center/>

<center><img src="https://noticiasdelaciencia.com/upload/images/02_2020/2236_a132-dj%C2%BDdj%C2%B5dj%C2%B9ndj%C2%BEdj%C2%BDdj%C2%BDndj%C2%B5-ndj%C2%B5ndj.png"><center/>

```
#reiniciar red neuronal
tensorflow.compat.v1.reset_default_graph()
#creamos una red neuronal #pasamos longitud de la fila
net=tflearn.input_data(shape=[None,len(training[0])])
#creamos las neuronas conectadas
#neuronas de entrada(input layers)
net=tflearn.fully_connected(net,10)
#neuronas intermedias(hidden layers)
net=tflearn.fully_connected(net,10)
#neuronas de salida(exit layers)
net=tflearn.fully_connected(net,len(exit[0]),activation="softmax",)
#aplicamos regresion a nuestra red
net=tflearn.regression(net, optimizer='adam', loss='categorical_crossentropy',)
model =tflearn.DNN(net)
print(len(all_words))
#moldeamos el modelo (n_epoch son las veces que se revisa la base de datos para obtener respuestas)
#bath_size son el # de entradas
#show metric muestrav el contenido
if os.path.isfile("/content/chatbot/Entrenamiento/model.tflearn.index"):
  model.load("/content/chatbot/Entrenamiento/model.tflearn")
else:
  model.fit(training,exit,n_epoch=10000,batch_size=3,show_metric=True)
  model.save("/content/chatbot/Entrenamiento/model.tflearn")
```

```
def response(update,context):
  bot=context.bot
  id=update.message.chat_id
  name=update.message.from_user.first_name
  texto=update.message.text
  if texto=="duerme":
    return f"HA SIDO UN GUSTO {name}, VUELVE PRONTO...."
  else:
    bucket=[0 for _ in range(len(all_words))]
    processed_sentence=nltk.word_tokenize(texto)
    processed_sentence=[stemmer.stem(palabra.lower()) for palabra in processed_sentence]
    for individual_word in processed_sentence:
      for i,palabra in enumerate(all_words):
        if palabra==individual_word:
          bucket[i]=1
    results=model.predict([numpy.array(bucket)])
    index_results=numpy.argmax(results)
    max=results[0][index_results]
    print(index_results)
    print(max)
    target=tags[index_results]
    print(tags)
    print(target)
    for tagAux in database['intents']:
      if tagAux['tag']==target:
         answer=tagAux['responses']
         answer=random.choice(answer)
    if max >= 0.75:
      if target=="Vuelos_disponibles":
        listAll(update,context)
      elif target=="vuelos_destino":
         searchd(update,context)
      elif target=="vuelos_origen":
         searcho(update,context)
      elif target=="ida":
         BUY_TICKET(update,context)
      elif target=="vuelta":
         BUYRT_TICKET(update,context)
      else:
        context.bot.sendMessage(chat_id=update.message.chat_id,text=answer)
    else:
      alternative="No entendi...\n"+"Escribe 'ayuda' para poder orientarte"
      context.bot.sendMessage(chat_id=update.message.chat_id,text=alternative)
```

```
def automatico(update,context):
  global ASIENTOS
  global ID
  global NAME
  print(VUELOS)
  bot=context.bot
  tamanio=len(ASIENTOS)
  print(tamanio)
  silla=ASIENTOS[tamanio-1]
  id=update.callback_query.message.chat_id
  name=NAME
  if int(silla) in ASIENTOS:
    bot.edit_message_media(
          chat_id=update.callback_query.message.chat_id,
          message_id=update.callback_query.message.message_id,
      media=InputMediaPhoto(
          media=open('/content/chatbot/agencia.PNG', 'rb'),
          caption=f'RESERVACION EXITOSA'
        )
      )
    nombre=name.lower()
    reservar(id,nombre,ID,silla)
    return ConversationHandler.END
  else:
    bot.edit_message_media(
          chat_id=update.callback_query.message.chat_id,
          message_id=update.callback_query.message.message_id,
      media=InputMediaPhoto(
          media=open('/content/chatbot/agencia.PNG', 'rb'),
          caption=f'RESERVACION FALLIDA.\n POSIBLEMENTE NO EXISTAN ASIENTOS DISPONIBLES!'
        )
      )

def asiento(update,context):
  bot=context.bot
  bot.edit_message_media(
          chat_id=update.callback_query.message.chat_id,
          message_id=update.callback_query.message.message_id,
      media=InputMediaPhoto(
          media=open('/content/photo.png', 'rb'),
          caption=f'ESCRIBE EL NUMERO DEL ASIENTO'
        )
      )
  return INPUT_ASIENTO
def input_asiento(update,context):
  global ASIENTOS
  global ID
  print(ASIENTOS)
  bot=context.bot
  silla=update.message.text
  chat=update.message.chat
  id=update.message.chat_id
  name=update.message.from_user.first_name
  try:
    if int(silla) in ASIENTOS:
      bot.send_message(
          chat_id=update.message.chat_id,
          text="RESERVACION REALIZADA CON EXITO !!\n"
      )
      nombre=name.lower()
      reservar(id,nombre,ID,silla)
      print("RESERVACION EXITOSA")
      return ConversationHandler.END
    else:
      bot.send_message(
          chat_id=update.message.chat_id,
          text="EL ASIENTO NO ESTA DISPONIBLE O NO EXISTE\n VUELVE A INTENTAR"
      )
      return ConversationHandler.END
  except:
    bot.send_message(
          chat_id=update.message.chat_id,
          text="ERROR:SOLO SE ACEPTA EL NUMERO DE ASIENTO\nVUELVA A INTENTAR!"
      )
```
